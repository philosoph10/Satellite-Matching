# Algorithm performance

First, let's look at the performance on different bands of the same image.

```python
image1_path = './images/T36UYA_20190805T083601_B02.jp2'
image2_path = './images/T36UYA_20190805T083601_B04.jp2'
```

```python
image1 = algo.read_image(image1_path)
image2 = algo.read_image(image2_path)
```

Now let's look at the output of the algorithm.

Since the images are already resized to 1,000 by 1,000

```python
# since the images are already resized i do not the target_size, but if the images have the original resilution 10,980 by 10,980,
# it is recommended to set it: target_size=(1000, 1000); also min_distance is set to 50 for demonstration purposes, for with
# smaller min_distance there would be too many keypoints; if no saving is required: save_to=None
_ = algo.detect_and_match_keypoints(image1, image2, min_distance=50, save_to='./different_bands.jpg')
```

As we can see, the lines are rather parallel to the axes, so the outputs are not random.
